```python
new_json = [] # Создаём пустой список для хранения итоговой структуры
for current_theme in themes:
    Theme = wc_data[wc_data['Theme'] == current_theme]['Theme'].iloc[0]                # Получаем значение соответствующей темы
    Cat = wc_data[wc_data['Theme'] == current_theme]['Cat'].iloc[0]                    # Получаем значение соответствующей категории
    WordCloud = list(wc_data[wc_data['Theme'] == current_theme]['WordCloud'].iloc[0])  # Получаем облака слов
    current_dict = {'Theme': Theme, 'Cat': Cat, 'WordCloud': WordCloud}                # Создаём текущий словарь
    new_json.append(current_dict)                                                      # Добавляем созданный словарь к списку
```
